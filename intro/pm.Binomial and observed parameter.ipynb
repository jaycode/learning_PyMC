# What is `observed` in pm.Binomial?

- The one with `observed` set does not have the posterior set for the observations. Not sure why though...
- `observed` value cannot be higher than `N`.
- `observed` can accept a list of values.


```python
import pymc as pm

upvotes = 10
downvotes = 10
N = upvotes + downvotes
upvote_ratio = upvotes/N
samples = 12

with pm.Model() as model:
    upvote_ratio = pm.Uniform("upvote_ratio", 0, 1)
    observations = pm.Binomial( "obs",  N, upvote_ratio, observed=[1, 2])

#     trace = pm.sample(samples, step=pm.Metropolis())
    trace = pm.sample(samples, step=pm.NUTS(),tune=int(samples/4),chains=1)
```
