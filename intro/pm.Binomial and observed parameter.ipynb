# What is `observed` in pm.Binomial?

1. The one with `observed` set does not have the posterior set for the observations. This is due to #4 below.
2. `observed` value cannot be higher than `N`.
3. `observed` can accept a list of values.
4. When `observed` is set, there is a new category `log_likelihood` whose `obs` has the shape of *[i, j]* Where *i* is the number of samples and *j* is the size of `observed`.

To conclude, by setting `observed`, we tell PyMC want to look at the 

```python
import pymc as pm

upvotes = 10
downvotes = 10
N = upvotes + downvotes
upvote_ratio = upvotes/N
samples = 12

with pm.Model() as model:
    upvote_ratio = pm.Uniform("upvote_ratio", 0, 1)
    observations = pm.Binomial( "obs",  N, upvote_ratio, observed=[upvotes])

#     trace = pm.sample(samples, step=pm.Metropolis())
    trace = pm.sample(samples, step=pm.NUTS(),tune=int(samples/4),chains=1)
```
